```python
minima,Q1,median,Q3,maxima=np.quantile(df['Insulin'],[0,0.25,0.50,0.75,1.0])
```

```python
IQR=Q3-Q1
lower_fence=Q1-1.5*(IQR)
higher_fence=Q3+1.5*(IQR)
```

```python

l=[]

for i in df['Insulin']:
    if i>=lower_fence and i<=higher_fence:
        l.append(i)
    else:
        df['Insulin']=df['Insulin'].replace(i,df['Insulin'].mean())
        
```
