### ANS 2 scatter plot is a plot which represent data in the points and in scatter point we can adjust the size,colour and intensity of the pont 

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
import matplotlib.pyplot as plt
plt.plot(x,y,'--b')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### ANS 3
subplot fucntion is used to show the plot and can be used to to show multiple parts of plots together.

```python
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])  
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])  
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])  
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])  
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1,'g')
plt.title('chart1')

plt.subplot(2,2,2)
plt.plot(x2,y2,'b')
plt.title('chart2')

plt.subplot(2,2,3)
plt.plot(x3,y3,'y')
plt.title('chart3')


plt.subplot(2,2,4)
plt.plot(x4,y4,'r')
plt.title('chart4')
            
plt.show()
```

### ANS 4
bar plot is a type of plot which is used to represent data in the form of bar. it is used for the visualization of the data

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.subplot(1,2,1)
plt.bar(company,profit)

plt.subplot(1,2,2)
plt.barh(company,profit)

plt.tight_layout()
plt.show()
```

### ANS 5
box plot is used to compare the data distribution of different categories.

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data=[box1,box2]

plt.boxplot(data,labels=['box1','box2'])


plt.xlabel('this is data')
plt.title('this is the box representation')
plt.show()
```
